```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mt
import seaborn as sns
```

```python
df=pd.read_csv(r"/Users/vishal69/Downloads/googleplaystore.csv")
```

```python
df.drop_duplicates(subset="App", inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(int)
```

```python
df['Size'] = df['Size'].apply(lambda x: float(x.strip('M').strip('k')) if 'M' in x or 'k' in x else 0)
```

```python
df['Reviews'] = df['Reviews'].astype(int)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.dropna(inplace=True)
```

```python
df['Price'] = df['Price'].apply(lambda x: float(x.replace('$', '')))
```

Replacing the Price Column to numeric

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

Analysis - The majority of apps have ratings between 4.0 and 5.0, indicating a positive angle in the distribution.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Size'], bins=30, kde=True)
plt.title("Distribution of App Sizes")
plt.xlabel("Size (in MB)")
plt.ylabel("Count")
plt.show()
```

Analysis - Most apps have sizes less than 50MB, with a few larger apps in the distribution.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Category', data=df, palette='viridis')
plt.xticks(rotation=90)
plt.title('Count of Apps by Category')
plt.show()
```

Analysis -  The 'Family' and 'Game' categories have the most apps on the Google Play Store.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.title("Distribution of App Prices")
plt.xlabel("Price (in USD)")
plt.ylabel("Count")
plt.show()
```

Analysis - The majority of apps on the Google Play Store are free, with a few premium apps priced above $50.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Rating', y='Category', data=df, alpha=1)
plt.title('Ratings vs. Category')
plt.show()
```

Analysis - Majority of the apps have a rating of just over 3.5 to 5, showing a positive review towards most of the apps.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Price', data=df, palette='Set3')
plt.xticks(rotation=90)
plt.title('App Prices by Category')
plt.show()
```

Analysis - The above data shows that most of the apps that have a price of over 300 USD are more focused towards the apps with the Genres of Medical, Lifestyle and Finance.

```python
plt.figure(figsize=(8, 6))
df['Content Rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Content Rating Distribution")
plt.ylabel("")
plt.show()
```

Analysis - Majority of the app content is for the age group "Everyone". Whereas just over 20% of the apps have an age criteria.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Rating', data=df)
plt.title("Price vs. Rating")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

Analysis - Majority of the apps are free to use with the second majority costing just under 50 USD and the rest over 300 USD, but the ratio of the ratings of the apps within this criteria has a difference with only the free apps having a rating of under 3, whereas the paid apps consistently have a rating above 3.

```python
plt.figure(figsize=(12, 6))
df['Genres'].value_counts().nlargest(40).plot(kind='bar')
plt.title("Most Apps in a Genre")
plt.xlabel("Genre")
plt.ylabel("Number of Apps")
plt.xticks(rotation=90)
plt.show()
```

Analysis - The apps with the majority in the following Genres are as follows with the majority being taken over by Tools, Entertainment, Education, Action. But the minority has been more focused towards the apps which provide info about Events, Card, Art/Design, Comics.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Size'], bins=10, kde=True)
plt.title("Distribution of App Sizes")
plt.xlabel("Size (MB)")
plt.ylabel("Frequency")
plt.show()
```

Analysis - Majority of the apps have a size of under 50MB, with most of the frequency focused towards the 30-60MB range.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df[df['Price'] > 0]['Price'], bins=30, kde=True)
plt.title("Price Distribution (Paid Apps)")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Analysis - Majority of the apps cost under the price range of 50 USD with only a snall set of apps that cost over it which is more focused towards the upper range of 300 USD.
